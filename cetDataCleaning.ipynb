```python
import json
import re

with open('cet.csv','r',errors='ignore') as f:
  fdata = f.read()
```

```python
datalist2 = fdata.split('\n')
```

```python
delimited = list(filter(lambda y:y!=None,map(lambda x:x.split(','),datalist2)))
```

```python
for z in range(len(delimited)):
  k = delimited[z]
  nl = []
  for j in k:
    if j=='':
      pass
    else:
      nl.append(j)
  delimited[z]=nl
```

```python
def is_number(string):
    try:
        float(string)  # or int(string)
        return True
    except ValueError:
        return False
l = ['1G', '1K', '1R', '2AG', '2AK', '2AR', '2BG', '2BK', '2BR', '3AG', '3AK', '3AR', '3BG', '3BK', '3BR', 'GM', 'GMK', 'GMR', 'SCG', 'SCK', 'SCR', 'STG', 'STK', 'STR']
index_GM = l.index('GM')
```

```python
def makeList(catind):
  pattern = r'^E\d{3}.+'
  pattern2 = r'^"E\d{3}.+'
  count = 0
  d = dict()
  list_of_d = []
  for lis in delimited:
    if len(lis)>0:
      if re.match(pattern,lis[0]):
        collegecode = lis[0][:4]
        collegename = lis[0][4:]
        list_of_d.append(d)
        d = dict()
        d['name'] = collegename
      elif re.match(pattern2,lis[0]):
        collegecode = lis[0][1:5]
        collegename = lis[0][5:]
        list_of_d.append(d)
        d = dict()
        d['name'] = collegename
      elif len(lis)==1:
        buffer = lis[0]
      elif len(lis)>(catind+1):
        if is_number(lis[catind+1]):
          branchname = buffer + lis[0]
          if branchname[0]=='\"':
            fbranchname = branchname[1:3]
          else:
            fbranchname = branchname[:2]
          buffer = ""
          gmcutoff = lis[catind+1]
          d[fbranchname] = gmcutoff
  return list_of_d[1:]
```

```python
list_of_dataList = []
for catindex in range(len(l)):
  res  = makeList(catindex)
  list_of_dataList.append(res)
print(len(list_of_dataList))
print(len(l))
```

```python
for datalis in range(len(list_of_dataList)):
  print(l[datalis])
  for items in range(10):
    print(list_of_dataList[datalis][items])
```

```python
for index_cat in range(len(l)):
  jsfilename = f'./jsfiles/{l[index_cat]}.json'
  with open(jsfilename,'w') as jsfp:
    json.dump(list_of_dataList[index_cat],jsfp)
```
